Порахуємо кількість дій кожного користувача у розрізі виду цієї дії

```python
users_actions_raw = df.groupby(["user_id", "event_name"]).agg(
    cnt_actions=('user_id', 'count'),
    coins=('coins', 'sum')
)
```

```python
users_actions_raw["cnt_actions"]
```

Тут будуємо каркас таблиці для визначення тих юзерів, що не зробили жодної події певного типу. Цим каркасом є scheme

```python
users = df['user_id'].drop_duplicates().reset_index(drop=True)
```

```python
scheme = users.to_frame().merge(df['event_name'].drop_duplicates().reset_index(drop=True), how='cross')
```

```python
users_actions = scheme.merge(users_actions_raw, how='left', on=["user_id", "event_name"])
users_actions.fillna(0, inplace=True)
```

Тут датафрейм freeloaders позначає юзерів, які не зробили жодної оплати в додатку, а has_paid - юзерів, що відкривали розділи книги за коіни.

```python
freeloaders = users_actions[(users_actions['event_name'] == 'purchase_success') & (users_actions['cnt_actions'] == 0)]["user_id"]
has_paid = users_actions[(((users_actions['event_name'] == 'click_unlock_now') | (users_actions['event_name'] == 'auto_unlock_chapter')) & (users_actions['cnt_actions'] != 0))]["user_id"]
```

```python
freeloaders = freeloaders.reset_index(drop=True).to_frame()
has_paid = has_paid.reset_index(drop=True).to_frame()
```

```python
has_paid
```

```python
freeloaders["is_freeloader"] = 1
has_paid["has_paid"] = 1
```

Додаємо в початковий датафрейм колонки, які позначають, що цю дію зробив юзер, який у наслідку виявився "халявщиком" і чи розблоковував він розділи за коіни.

Рахуємо кількість книжок, які прочитав кожен користувач

```python
books_per_user = df.groupby(["user_id"], as_index=False).agg(
    cnt_books=('book_id', 'nunique')
)
```

Кожному юзеру в початковому датафреймі проставляємо у відповідність кількість книжок, які він прочитав.

```python
avg_books_per_user = df.merge(books_per_user, how='left', on='user_id')
```

```python
avg_books_per_user
```

Рахуємо середню кількість книжок, які прочитали читачі цієї книги.

```python
avg_books_per_user = avg_books_per_user.groupby(["book_id", "user_id"]).agg(
    cnt_books=('cnt_books', 'min')
).groupby("book_id").agg(
    cnt_books=('cnt_books', 'mean')
)
```

```python
avg_books_per_user
```

Тут possible_fraudsters - це дані про дії користувачів, які і не здійснювали оплат, і розблоковували платні розділи в книжках

```python
possible_fraudsters = df[(df["is_freeloader"] == 1) & (df["has_paid"] == 1)]
```

```python
freeloaders_donaters_on_books = possible_fraudsters.groupby(["book_id", "event_name"], as_index=False).agg(
    donates_from_freeloaders=("user_id", "nunique")
)
```

Порахували кількість безоплатних донатерів на кожній книжці

```python
freeloaders_donates_on_books = freeloaders_donaters_on_books[freeloaders_donaters_on_books["event_name"] == 'click_send_donate'][["book_id", "donates_from_freeloaders"]]
```

```python
freeloaders_donates_on_books
```

```python
final_book_df = df.groupby(["book_id", "user_id"], as_index=False).agg(
    has_paid=("has_paid", "min"),
    is_freeloader=("is_freeloader", "min")
)
final_book_df["both"] = final_book_df["has_paid"] * final_book_df["is_freeloader"]
```

```python
final_book_df
```

```python
final_book_df = final_book_df.groupby("book_id").agg(
    readers=("user_id", "count"),
    both=("has_paid", "sum")
)
```

```python
final_book_df
```

```python
final_book_df["freeloaders_paid_for_chapters_percentage"] = final_book_df["both"] / final_book_df["readers"]
```

```python
final_book_df
```

Ітак, маємо готовий каркас для додавання важливих для виявлення фроду фіч по кожній книжці. Залишилося додати середній час "проклікування" розділів, кількість донатерів-халявщиків і середню кількість книжок на одного читача цієї книжки.

```python
time_df = pd.read_csv("F:/Кейс 3/time_on_chapter.csv")
```

```python
time_df
```

```python
time_df.describe()
```

Бачимо, що середній час перегляду розділу сильно викривлений через викиди в даних, а саме час перегляду розділу юзерами, які забули про додаток на значний час (місяць, рік). Тому дивлячись на міру середнього для часу, який юзери витратили на розділи конкретної книги, будемо брати медіану, а не середнє, оскільки вона стійкіша до викидів.

```python
median_time_reading = time_df.groupby(["book_id", "user_id"]).agg(
    median_time_reading=("time_reading", "median")
)
```

У розрізі ж кожної книжки будемо брати середнє арифметичне часу, який юзери витратили на розділи книги. Так зможемо оцінити, на яких книжках було суттєво багато "проклікувань" розділів. Але треба прийняти до уваги, що такий підхід може зачепити книжки, в яких розділи самі по собі невеликі.

*звичайно ж, правильніше було б дивитися в розрізі саме "халявщиків", але на жаль я це занадто пізно зрозумів(

```python
average_time_reading = median_time_reading.groupby("book_id").agg(
    average_time_reading=("median_time_reading", "mean")
)
```

```python
average_time_reading
```

```python
final_book_df = final_book_df.merge(average_time_reading, on="book_id")
final_book_df = final_book_df.merge(avg_books_per_user, on="book_id")
final_book_df = final_book_df.merge(freeloaders_donates_on_books, on="book_id", how="left")
final_book_df.fillna(0, inplace=True)
```
